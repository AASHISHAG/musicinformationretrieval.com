# K-Means Clustering

```python
from sklearn import preprocessing
from sklearn.cluster import KMeans, AffinityPropagation
import essentia.standard as ess
```

Sometimes, an unsupervised learning technique is preferred.  Perhaps you do not have access to adequate training data, or perhaps the training data's labels are not completely clear.  Maybe you just want to quickly sort real-world, unseen, data into groups based on its feature similarity.

In such cases, clustering is a great option!

Let's try clustering with a familiar bunch of audio files and code.

Download `simpleLoop.wav`, and play it.

```python
from urllib import urlretrieve
urlretrieve('https://ccrma.stanford.edu/workshops/mir2014/audio/simpleLoop.wav', filename='simpleLoop.wav')

from IPython.display import Audio
Audio('simpleLoop.wav')
```

Load the audio file into an array.

```python
simple_loop = ess.MonoLoader(filename='simpleLoop.wav')()
```

## Feature Extraction

Let's compute the zero crossing rate and energy for each frame in the audio signal.

```python
zcr = ess.ZeroCrossingRate()
energy = ess.Energy()

def compute_features(frame):
    return array( [zcr(frame), energy(frame)] )
```

```python
features = array([compute_features(frame) for frame in ess.FrameGenerator(simple_loop, frameSize=1024, hopSize=500)])
print features.shape
```

```python
plot(simple_loop)
```

Plot the zero crossing rate:

```python
plot(features[:,0])
ylabel('Zero Crossing Rate')
```

Plot the energy:

```python
plot(features[:,1])
ylabel('Energy')
```

## Feature Scaling

Scale the features (using the scale function) from -1 to 1.

```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1))
features_scaled = min_max_scaler.fit_transform(features)
print features_scaled.shape
print features_scaled.min(axis=0)
print features_scaled.max(axis=0)
```

Plot the features.

```python
scatter(features_scaled[:,0], features_scaled[:,1])
xlabel('Zero Crossing Rate (scaled)')
ylabel('Energy (scaled)')
```

## Using K-Means

Time to cluster! Let's initialize the algorithm to find three clusters.

```python
model = KMeans(n_clusters=3)
labels = model.fit_predict(features_scaled)
features_scaled.shape
print labels
```

Plot the results.

```python
scatter(features_scaled[labels==1,0], features_scaled[labels==1,1], c='r')
scatter(features_scaled[labels==2,0], features_scaled[labels==2,1], c='y')
scatter(features_scaled[labels==0,0], features_scaled[labels==0,1], c='k')
xlabel('Zero Crossing Rate (scaled)')
ylabel('Energy (scaled)')
```

## Affinity Propagation

In scikit-learn, other clustering algorithms such as *affinity propagation* can cluster without defining the number of clusters beforehand.

```python
fs = 44100
loader = ess.MonoLoader(filename='../test/beatbox_jason.wav')
audio = loader()
```

Get onset times:

```python
get_onsets = ess.OnsetRate()
onset_times, onset_rate = get_onsets(audio)
print onset_times
```

```python
onset_samples = [int(fs*i) for i in onset_times]
print onset_samples
```

```python
frame_sz = int(0.200*fs)
segments = array([audio[i:i+frame_sz] for i in onset_samples])
```

```python
features = array([compute_features(segment) for segment in segments])
print features.shape
```

Scale the features:

```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1))
features_scaled = min_max_scaler.fit_transform(features)
```

Get cluster labels:

```python
#model = KMeans(n_clusters=3)
model = AffinityPropagation(damping=0.5)
labels = model.fit_predict(features_scaled)
features_scaled.shape
print labels
```

Plot features:

```python
scatter(features_scaled[labels==0,0], features_scaled[labels==0,1], c='k')
scatter(features_scaled[labels==1,0], features_scaled[labels==1,1], c='r')
scatter(features_scaled[labels==2,0], features_scaled[labels==2,1], c='y')
scatter(features_scaled[labels==3,0], features_scaled[labels==3,1], c='b')
scatter(features_scaled[labels==4,0], features_scaled[labels==4,1], c='g')
scatter(features_scaled[labels==5,0], features_scaled[labels==5,1], c='m')
xlabel('Zero Crossing Rate (scaled)')
ylabel('Energy (scaled)')
```

Play a beep upon each frame in the same cluster:

```python
onsets_marker = ess.AudioOnsetsMarker(onsets=onset_times[labels==1], type='beep')
with_beeps = onsets_marker(audio)
Audio(data=with_beeps, rate=fs)
```
