# Segmentation

In audio processing, it is common to operate on one frame at a time. Frames are typically chosen to be 10 to 100 ms in duration.

For this workshop, we will use `essentia.standard.FrameGenerator` to segment our audio signal.

Let's create an audio sweep signal that is frequency modulated from 110 Hz to 880 Hz:

```python
import essentia
T = 3.0
fs = 44100.0
t = linspace(0, T, T*fs, endpoint=False)
f0 = 440*logspace(-2, 1, T*fs, endpoint=False, base=2.0)
x = essentia.array(sin(2*pi*f0*t))
print f0.min(), f0.max()
```

Listen to the signal:

```python
from IPython.display import Audio
Audio(x, rate=fs)
```

## Example: Zero Crossing Rate

For each frame, compute the zero crossing rate, and display:

```python
from essentia.standard import FrameGenerator, ZeroCrossingRate
zcr = ZeroCrossingRate()
semilogy([zcr(frame) for frame in FrameGenerator(x, frameSize=1024, hopSize=512)])
```

## Example: MFCCs

For each frame, compute the mel frequency cepstral coefficients (MFCCs), and display:

```python
from essentia.standard import MFCC, Spectrum, Windowing, FrameGenerator
hamming_window = Windowing(type='hamming')
spectrum = Spectrum()  # FFT() would return the complex FFT, here we just want the magnitude spectrum
mfcc = MFCC()

mfccs = array([mfcc(spectrum(hamming_window(frame)))[1]
               for frame in FrameGenerator(x, frameSize=1024, hopSize=512)])

imshow(mfccs[1:,:].T, origin='lower', aspect='auto')
ylabel('MFCC Coefficient Index')
xlabel('Frame Index')
```
